```python
import numpy as np
import pandas as pd
import pymc3 as pm
import daft
```

```python
total_no_of_balls = 10
no_of_red = 8
no_of_white_ball = 2

model = pm.Model();
    
with model:
    red_ball_more_than_7_times = pm.Binomial('red_ball_8_times',n=8,p=(no_of_red/total_no_of_balls))
    red_ball_comes_6_times = pm.Binomial('red_ball_comes_6_times',n=8,p=((no_of_red-2)/total_no_of_balls))
    
    trace = pm.sample(100000,tune=5000,progressbar=True)   
```
